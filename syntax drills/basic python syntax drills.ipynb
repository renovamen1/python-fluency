```python
#dict ma iterate garekoo ho
fruits = {'apple':'250','banana':'60','pear':'300',"orange":"200"}
for fruit,price in fruits.items():
    print(f'the current price of {fruit} is {price}')
```

```python
# this is dict comprehension
test = {fruit: int(price)*2 for fruit,price in fruits.items()}
test
```

```python
#dict comprehension with conditionals
pricing = {'claude':200,'grok':100,"chatgpt":150,'deepseek':25}
test3 ={model:price*2 for model,price in pricing.items() if price<=100}
test3
test4 ={model:price-30 for model,price in pricing.items()if model == 'claude' or model == 'chatgpt'}
test4
```
